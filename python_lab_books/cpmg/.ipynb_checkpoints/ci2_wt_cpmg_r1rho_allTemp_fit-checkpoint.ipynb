```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import os
import math
import seaborn as sns
from scipy import optimize
from scipy.optimize import minimize
```

```python
# Using readline() 
path1='/Users/psz898/OneDrive/Documents/SCIENCE/UCPH/info_comp/cpmg_for_korznev/cpmg_fit/ci2/'
systemf = 'backups_wt/'


file_par = open(f'{path1}{systemf}fit_allr_allt_2st_ord2.par')
count = 0
temp = []
pA = []
pA_err = []
kAB = []
kAB_err = []
cs_res = []
cs = []
cs_err = []

regexPA = re.compile("[PA:|PB:]\s+(\d+)\s+(\d.\d+)\s+(\d.\d+)\s+\w\s*")
# PA:         5   0.957329   0.000000   f
regexKAB = re.compile("KAB:\s+\d+\s+(\d+.\d+)\s+(\d+.\d+)\s+\w\s*")

regexCS = re.compile("CS\d_[A\|B]:\s+(\d+)\s+N\s+\d\s+(-?\d+.\d+)\s+(\d+.\d+)\s+\w\s*")



while True: 
    count += 1
  
    # Get next line from file 
    line = file_par.readline() 
  
    # if line is empty 
    # end of file is reached 
    if not line: 
        break
    resultPA = regexPA.search(line)
    resultKAB = regexKAB.search(line)
    resultCS = regexCS.search(line)

    
    #print("Line{}: {}".format(count, line.strip())) 
    if resultPA:
        #print (result.group(1), result.group(2), result.group(3))
        temp.append(int(resultPA.group(1)))
        pA.append(float(resultPA.group(2)))
        pA_err.append(float(resultPA.group(3)))
     
    if resultKAB:
        #print (resultKAB.group(1), resultKAB.group(2))
        kAB.append(float(resultKAB.group(1)))
        kAB_err.append(float(resultKAB.group(2)))
    
    if resultCS:
        #print (resultCS.group(1), resultCS.group(2), resultCS.group(3))
        cs_res.append(float(resultCS.group(1)))
        cs.append(float(resultCS.group(2)))
        cs_err.append(float(resultCS.group(3)))

    
    
file_par.close() 

rows, cols = (3, 1) 
PA = [[0 for i in range(cols)] for j in range(rows)]
KAB = [[0 for i in range(cols)] for j in range(rows)]
CS = [[0 for i in range(cols)] for j in range(rows)]

kelvin=273.15
celsius=0
t_type=celsius

PA[0]=[x+t_type for x in temp[1::2]]

PA[1]=[x*100 for x in pA[0::2]]
PA[2]=[x*100 for x in pA_err[1::2]]

PAtr= np.array(PA).T

KAB[0]=[x+t_type for x in temp[1::2]] 
KAB[1]=kAB
KAB[2]=kAB_err
KABtr= np.array(KAB).T

CS[0] = cs_res
CS[1] = cs
CS[2] = cs_err
CStr= np.array(CS).T


df_pA = pd.DataFrame(PAtr, columns=["T", "pA", "pA_err"])
df_pA['T']= df_pA['T'].astype(int)
#df_pA

df_KAB = pd.DataFrame(KABtr, columns=["T", "Kex", "KAB_err"])
df_KAB['T']= df_KAB['T'].astype(int)
# df_KAB

df_CS = pd.DataFrame(CStr, columns=["res", "A-B", "CS_err"])
#df_CS

df_pA
```

```python
# cs=c0+c1*t+c2*t^2
def myfunct(x,t):
    x = np.asarray(x)
    return x[1] + x[2]*t + x[3]*t**2

# def myfuncterr(x,t):
#     x = np.asarray(x)
#     return np.sqrt((x[1])**2 + (x[2]*t)**2 + (x[3]*t**2)**2)
def myfuncterr(x,t):
    x = np.asarray(x)
    return np.sqrt((x[1])**2 + (x[2])**2 + (x[3])**2)

df_CS_25 = df_CS.groupby('res')['A-B'].agg(lambda x: myfunct(x,25))
df_CS_25err = df_CS.groupby('res')['CS_err'].agg(lambda x: myfuncterr(x,25))


df_CS_25 = pd.DataFrame(df_CS_25).reset_index()
df_CS_25err = pd.DataFrame(df_CS_25err).reset_index()
df_CS_25['res']= df_CS_25['res'].astype(int)
df_CS_25['A-Berr']= df_CS_25err['CS_err']
df_CS_25['type'] = 'A-B' 

df_CS_5_25 = pd.DataFrame()

df_CS_5_25['res'] = df_CS_25['res']
df_CS_5_25['A-B'] = [0.34,0.13,0.7,0.34,0.35,0.07]
df_CS_5_25['A-Berr'] = [0.34*0.1,0.13*0.1,0.7*0.1,0.34*0.1,0.35*0.1,0.07*0.1]
df_CS_5_25['type'] = '5-25' 


df_dCS_AB_temp = df_CS_25.append(df_CS_5_25)


df_dCS_AB_temp
```

```python
def get_cs_AB_atTemp(temp):
    # cs=c0+c1*t+c2*t^2
    def myfunct(x,t):
        x = np.asarray(x)
        return x[1] + x[2]*t + x[3]*t**2
    
    # def myfuncterr(x,t):
    #     x = np.asarray(x)
    #     return np.sqrt((x[1])**2 + (x[2]*t)**2 + (x[3]*t**2)**2)
    def myfuncterr(x,t):
        x = np.asarray(x)
        return np.sqrt((x[1])**2 + (x[2])**2 + (x[3])**2)
    
    df_CS_X = df_CS.groupby('res')['A-B'].agg(lambda x: myfunct(x,temp))
    df_CS_Xerr = df_CS.groupby('res')['CS_err'].agg(lambda x: myfuncterr(x,temp))
    
    
    df_CS_X = pd.DataFrame(df_CS_X).reset_index()
    df_CS_Xerr = pd.DataFrame(df_CS_Xerr).reset_index()
    df_CS_X['res']= df_CS_X['res'].astype(int)
    df_CS_X['A-Berr']= df_CS_Xerr['CS_err']
    df_CS_X['temp'] = temp 
    
#     df_CS_5_25 = pd.DataFrame()
    
#     df_CS_5_25['res'] = df_CS_25['res']
#     df_CS_5_25['A-B'] = [0.34,0.13,0.7,0.34,0.35,0.07]
#     df_CS_5_25['A-Berr'] = [0.34*0.1,0.13*0.1,0.7*0.1,0.34*0.1,0.35*0.1,0.07*0.1]
#     df_CS_5_25['type'] = '5-25' 
    
#     df_CS_X = df_CS_X.append(df_CS_5_X)
    
    return df_CS_X

df_CS_5 = get_cs_AB_atTemp(5)
df_CS_10 = get_cs_AB_atTemp(10)
df_CS_15 = get_cs_AB_atTemp(15)
df_CS_20 = get_cs_AB_atTemp(20)
df_CS_25 = get_cs_AB_atTemp(25)

df_all = [df_CS_5,df_CS_10,df_CS_15,df_CS_20,df_CS_25]
df = pd.concat(df_all).iloc[0::1].reset_index()
df_CS=df.drop(columns=['index'])
df_CS
```

```python
colors=['#226894', '#e47632', '#1e873a', '#c03439', '#8c67a6']

def grouped_barplot(df, cat,subcat, val , err):
    fig, axs = plt.subplots(ncols=1,figsize=(2.5,3),dpi=400)
    u = df[cat].unique()
    x = np.arange(len(u))
    subx = df[subcat].unique()
    offsets = (np.arange(len(subx))-np.arange(len(subx)).mean())/(len(subx)+1.)
    width= np.diff(offsets).mean()
    for i,gr in enumerate(subx):
        dfg = df[df[subcat] == gr]
        axs.bar(x+offsets[i], dfg[val].values, width=width, 
                label="{} {}".format(subcat, gr), yerr=dfg[err].values, color=colors[i])
    #plt.xlabel(cat)

    plt.xlabel("Residues")    
    #plt.ylim([0,8])
    plt.ylabel('ppm')
    plt.xticks(x, u)
    #plt.title("CSs")
    #plt.title( '$ S^2_{axis} $ @' + str(field) +' MHz')#, fontsize = 20
    #plt.title( '$ S^2_{axis} $ @' + str(field) +' MHz')
    plt.legend(['|\u0394\u03C9|','\u0394CS'],ncol=1,prop={"size":10})
    #plt.show()
    #plt.yticks([0,10,20,30]);
    xticks = ['K24','E26','M40','I44','D45','V47']
    plt.xticks(fontsize=10, rotation=45)
    axs.set_xticklabels(xticks);
    plt.tight_layout()
    
    plt.savefig(f'CI2_WT_allTemp_CS.pdf');

    
    
cat = "res"
subcat = "type"
val = "A-B"
err = "A-Berr"
field=600;
grouped_barplot(df_dCS_AB_temp, cat, subcat, val, err )
```
