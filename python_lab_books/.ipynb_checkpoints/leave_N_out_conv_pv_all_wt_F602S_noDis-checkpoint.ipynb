```python
def binning (min_bin,max_bin,step, df):
    #step =10
    bins=range(min_bin,max_bin+step,step)
    y, bins_arr = np.histogram(df[:].values, bins)
    y_norm = (y/y.sum())*100
    return y, y_norm, bins_arr
```

```python
def leave_N_out(df, nblocks):

    ######
    
    i = 0;
    block_val = [];
    step = m.floor(df.shape[0]/nblocks)

    for i in range(1,df.shape[0],step):
        block_val.append(df[:].iloc[i:i+step].values.mean())  
    

    block_val_np = np.array(block_val)
    #print(block_val_np)
    ######
    
    blocks_idx_out = np.arange(0,nblocks,1)
    
    deltas = []
    for i in range(1, int(nblocks/2)+1 ):
        combs = combinations(blocks_idx_out,i)
      
        delta=[]
        for acomb in combs:
            acombl = list (acomb)
            
            mask1 = np.ones(nblocks)
            mask2 = np.zeros(nblocks)
            
            mask1[acombl] = 0
            mask2[acombl] = 1
            
            data_main =  np.average(block_val_np.compress( mask1, axis = 0 ))
            data_out  =  np.average(block_val_np.compress( mask2, axis = 0 ))
        
            tmp = abs(data_main-data_out)
            delta.append(tmp)
        
        deltas.append( [np.average(delta), np.std(delta)/np.sqrt(len(delta))] )
    
    deltas = np.array(deltas)

    return deltas
```

```python
nblocks=20
deltas_cort_wt_dt1000 = leave_N_out(df_cort_wt_dt1000, nblocks)
deltas_pred_wt_dt1000  = leave_N_out(df_pred_wt_dt1000, nblocks)
deltas_dex_wt_dt1000   = leave_N_out(df_dex_wt_dt1000, nblocks)
deltas_dibC_wt_dt1000  = leave_N_out(df_dibC_wt_dt1000, nblocks)

deltas_cort_F602S_dt1000 = leave_N_out(df_cort_F602S_dt1000, nblocks)
deltas_pred_F602S_dt1000  = leave_N_out(df_pred_F602S_dt1000, nblocks)
deltas_dex_F602S_dt1000   = leave_N_out(df_dex_F602S_dt1000, nblocks)
deltas_dibC_F602S_dt1000  = leave_N_out(df_dibC_F602S_dt1000, nblocks)
```
