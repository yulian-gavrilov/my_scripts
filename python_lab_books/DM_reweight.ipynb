```python
import numpy as np
import pandas as pd
import math as m
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcol
import matplotlib.cm as cm
```

```python
def reweight_SDM(system):
    with open(f'{system}_CA_dt100_24000fr_240bl.npy', 'rb') as f:
        stdm = np.load(f)
        
    weights = np.load(f'./results_rw_{system}.pkl',allow_pickle=True)
    w = np.concatenate((weights[300][20:100], weights[300][120:200], weights[300][220:300]), axis=None)
    
    stdm_rw = np.average( stdm, weights = w, axis = 0)  
    #stdm_rw = np.average( stdm, axis = 0)  
    
    return stdm_rw
```

```python
stdm_rw_wt_disp = reweight_SDM('wt_disp')
stdm_rw_l49i_disp = reweight_SDM('l49i_disp')
stdm_rw_i57v_disp = reweight_SDM('i57v_disp')
stdm_rw_l49i_i57v_a_disp = reweight_SDM('l49i_i57v_a_disp')
stdm_rw_l49i_i57v_b_disp = reweight_SDM('l49i_i57v_b_disp')
```

```python
delta_l49i_disp = stdm_rw_l49i_disp-stdm_rw_wt_disp
delta_i57v_disp = stdm_rw_i57v_disp-stdm_rw_wt_disp
delta_l49i_i57v_a_disp = stdm_rw_l49i_i57v_a_disp-stdm_rw_wt_disp
delta_l49i_i57v_b_disp = stdm_rw_l49i_i57v_b_disp-stdm_rw_wt_disp
```

```python
delta_l49i_disp_sum = round(sum(sum(delta_l49i_disp))/2,2)
delta_i57v_disp_sum = round(sum(sum(delta_i57v_disp))/2,2)
delta_l49i_i57v_a_disp_sum = round(sum(sum(delta_l49i_i57v_a_disp))/2,2)
delta_l49i_i57v_b_disp_sum = round(sum(sum(delta_l49i_i57v_b_disp))/2,2)

print("System \t\t\ttotal sum")
print("L49I-WT (disp)\t\t ",delta_l49i_disp_sum)
print("57V-WT (disp)\t\t",delta_i57v_disp_sum)
print("L49I_I57V_A-WT (disp)\t",delta_l49i_i57v_a_disp_sum)
print("L49I_I57V_B-WT (disp)\t",delta_l49i_i57v_b_disp_sum)
```
